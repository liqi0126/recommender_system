## Get wiki data

```python
import wptools
import pandas as pd

def get_wiki_pages_by_cat(cat_name, pages, titles, cates, depth=2):
    cat = wptools.category(cat_name)
    cat_members = cat.get_members(show=False)
    if 'members' in cat_members.data.keys():
        for member in cat_members.data['members']:
            title = member['title']
            if title not in titles:
                titles.append(title)
                pages.append([title, wptools.page(title, silent=True).get_parse().data['wikitext']])
    
    if depth > 0 and 'subcategories' in cat_members.data.keys():
        subs = cat_members.data['subcategories']
        for sub in subs:
            cate = sub['title']
            if cate not in cates:
                cates.append(cate)
                get_wiki_pages_by_cat(cate, pages, titles, cates, depth-1)
```

```python
pages = []
titles = []
cates = []
get_wiki_pages_by_cat('Category:World War II', pages, titles, cates, depth=100)
```

```python
df = pd.DataFrame(pages, columns=['title', 'text'])
```

```python
df.to_csv('ww2.csv', index=False)
```
